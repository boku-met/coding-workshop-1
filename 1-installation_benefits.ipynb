## Installable code can be used independent of the source location

* reuse code across your project without copying
* reuse code from other projects
* share code with colleagues

**coming up: 3 part example**

1. not installed but in the working directory
2. not installed and outside scope
3. installed and outside scope

*1*  
You can import code from modules (python files) in the current working directory.

```python
from noninstallable_module_example1 import hello
hello()
```

*2*  
However, you cannot do the same with files in subdirectories (without
adding their path to the Python search path).

```python
from noninstallable_module_example2 import hello
hello()
```

*3*  
This is different after installation of some module. Then, you can
import the code whenever the environment it has been installed to is
active.

```python
%pip install ./example_installable_external_resource
```

```python
from example_installable_external_resource import hello
hello()
```

```python
%pip uninstall -y example_installable_external_resource
```
