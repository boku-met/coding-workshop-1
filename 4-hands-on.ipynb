# Hands-on: make your code installable

First try to build a tiny library for the sake of this exercise, install
it, and test that you can import it. If that was successful, prepare a
function that you would like to share for a follow-up hands-on.

## Initiate library

Create a folder at a location of your choice. Then, create the necessary
files and structure inside. If you don't know or are unsure about what
this implies, use the following command inside the new folder to create
a Python library with the name of the working directory.

```shell
uv init --lib
```

## Create test function

While `uv` will create a dummy function that can be used to test the
library's installation, create an own test function in a new module to
convince yourself that you could have done this on your own.

Create a Python file in the directory that your library's `__init__.py`
resides in. There, define some function that prints something or returns
anything that can be represented where you test the installation (e.g.,
a notebook). If you wish, use a function of a package that is not
shipped with Python, e.g., `pandas`, to exercise declaring dependencies,
which is relevant in praxis. In this case, add the dependency (the
package from which you `import` the additional function) to the
`pyproject.toml`. If you're unsure about this, use `uv add --raw <pypi
package name>`.
